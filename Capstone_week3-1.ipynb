## Download and Explore Dataset

```python
#Store the url
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
# Create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html, "lxml")
type(soup)
```

```python
#Get the text
text = soup.get_text()
#print(text)
```

From these object extract the table data

```python
# Print the first 10 rows 
rows = soup.find_all('tr')
print(rows[:10])
```

```python
#Create a loop that iterate through table rows and print out the cells of the rows
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
#Use beautiful soup to to remove the html tags
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(data = list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

Lets select the first 3 columns and remove the first row with no value

```python
df1 = df1.ix[1:, :2]
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
df1[2] = df1[2].str.strip('\n]')
df1.head(10)
```

```python
df1.columns = ["Postcode", "Borough", "Neighborhood"]
```

```python
df1.head()
```

```python
df=df1[df1['Borough']!='Not assigned']
```

```python
df1.loc[df1['Neighborhood']=='Not assigned', ['Neighborhood']]=df1['Borough']
```

```python
geodata = pd.read_csv("https://cocl.us/Geospatial_data")
geodata.rename(columns = {'Postal Code' : 'Postcode'}, inplace = True)
geodata.head()
```

```python
#Merge dataframes
df= pd.merge(df1,geodata,
                 on='Postcode', 
                 how='left')
df = df.dropna()
```

```python
df.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df['Borough'].unique()),
        df.shape[0]
    )
)
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

Create a map of Toronto

```python
# create map of Toronto using latitude and longitude values
map_tr = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_tr)  
    
map_tr
```
